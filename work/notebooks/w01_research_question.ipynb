## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

The decision this improves: which pages from the backlog get picked up for content review this cycle, out of many more candidates than the team has time to touch.
Who acts on it: whoever owns the content/SEO review queue — they take the ranked list and work through it top-down instead of guessing or going alphabetically.
Cost of a wrong call: two ways to be wrong, and they cost differently.

False positive (recommending a page that didn't need work): wastes review time on a page that wouldn't have moved, while a page that actually needed help sits untouched.
False negative (missing a page that did need work): the page keeps losing traffic silently until someone notices, by which point recovery is harder and slower.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print(f"Rows: {len(df):,}")
print(f"Columns: {len(df.columns)}")
print()
print("All column names:")
print(df.columns.tolist())
print()
print("-" * 50)
print()

# client count — helps show scale/diversity of the data
if 'client_id' in df.columns:
    print(f"Unique clients: {df['client_id'].nunique()}")
else:
    print("client_id column not found — check name above")

print()

# trend direction breakdown — the core "needs review" signal
if 'trend_direction' in df.columns:
    print("trend_direction breakdown (%):")
    print(df['trend_direction'].value_counts(normalize=True).mul(100).round(1))
else:
    print("trend_direction column not found — check name above")

print()

# magnitude of decline for pages trending down
trend_pct_candidates = [c for c in df.columns if 'trend' in c.lower() and 'pct' in c.lower()]
print(f"Columns that look like trend-percent: {trend_pct_candidates}")

if 'trend_direction' in df.columns and trend_pct_candidates:
    pct_col = trend_pct_candidates[0]
    down_pages = df[df['trend_direction'] == 'down']
    print(f"\nUsing '{pct_col}' as the trend-percent column")
    print(f"Declining pages count: {len(down_pages):,}")
    print(f"Median {pct_col} for declining pages: {down_pages[pct_col].median():.1f}%")
```

## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

What this work will be able to say:
- Observed: which pages, in this anonymized sample, are trending down and by
  how much.
- Directional: which signals (position, CTR, freshness) tend to move together
  with declining pages — not proof that one causes the other.
- Decision-support: a ranked shortlist that helps a reviewer prioritize their
  time, not a replacement for their judgment on any single page.

What it will never say:
- Causal proof — that changing X on a page will cause its ranking to improve.
  Correlation in this data isn't evidence of what Google's algorithm actually
  weighs.
- "Predicting Google" — this model predicts patterns in *this client's*
  historical search data, not the search engine's future behavior or updates.
